# Set up Monkeylearn

The first thing to be done is to set your Monkeylearn credentials. You can [signup with MonkeyLearn](http://www.monkeylearn.com/) and get your [API token](https://app.monkeylearn.com/accounts/user/settings/) if you don't have one already.

```python
# Add your Monkeylearn Access Token here
API_KEY = ''
```

Here we will be defining some classifiers and extractors we will be talking to, the [News Categorizer](https://app.monkeylearn.com/categorizer/projects/cl_hS9wMk9y/), the [Entity Extractor](https://app.monkeylearn.com/extraction/extractors/ex_isnnZRbS/) and the  [Keyword Extractor](https://app.monkeylearn.com/extraction/extractors/ex_y7BPYzNG/). We will be using batch endpoints for each of them.

```python
# These are the URLs of the Classifiers and Extractors we will be using.
NEWS_CLASSIFIER = "https://api.monkeylearn.com/api/v1/categorizer/cl_hS9wMk9y/classify_batch_text/"
ENTITY_EXTRACTOR = "https://api.monkeylearn.com/api/v1/extraction/ex_isnnZRbS/extract_batch_text/"
KEYWORD_EXTRACTOR = "https://api.monkeylearn.com/api/v1/extraction/ex_y7BPYzNG/extract_batch_text/"
# We will send news articles in batches of 50.
BATCH_SIZE = 50
```

# Load and classify the data

We will be reading our data from a file containing a JSON object.

```python
import json

OUTLETS = ['cnn', 'fox', 'china_daily', 'bbc', 'al_jazeera']
# This is the name of the file that contains our news. Change it accordingly
# For example: cnn.json
#              al_jazeera.json etc.
src = 'cnn.json'

with open(src, 'rb') as src_f:
    src_data = json.load(src_f)['news']
    # We will keep a copy of the data to use later
    objects = src_data
```

Now let's classify all the raw data we have

```python
import requests

# We will store all of our results here. The first list will contain news
# classification, the second extracted entities, and the third will hold
# any extracted keywords. Positions will correspond to the position of the
# original piece of news in the objects variable.
partials = [[], [], []]
while src_data:
    # We will classify data in chunks of BATCH_SIZE.
    items = src_data[:BATCH_SIZE]

    # Prepare the data and headers for our request
    data = {'text_list': items}
    headers = {'Authorization': 'Token {0}'.format(API_KEY),
               'Content-Type': 'application/json'}

    # We will call the classifiers and extractors here, and then add the
    # result to the partials list.
    for i, url in enumerate([NEWS_CLASSIFIER, ENTITY_EXTRACTOR, KEYWORD_EXTRACTOR]):
        response = requests.post(url,
                                 data=json.dumps(data),
                                 headers=headers)
        result = json.loads(response.text)['result']
        partials[i] += result

    # Remove the classified items from the list of items.
    src_data = src_data[BATCH_SIZE:]

# Lets zip the original news with the data Monkeylearn returned. We will 
# end up with a list containing tuples that look like:
#   (news text, news categories, extracted entities, extracted keywords)
data = zip(objects, *partials)
```

# Processing time!

First of all, let's define some useful functions we will be using later

```python
# Gets the most specific category a piece of news is in.
def get_category(news):
    return map(lambda x: x['label'], news[1])[-1]

# Gets all the entities and their type of a piece of news.
def get_entities(news):
    return map(lambda x: (x['entity'], x['tag']), news[2])

# Gets the list of keywords associated to a piece of news.
def get_keywords(news):
    return map(lambda x: x['keyword'], news[3])
```

First, lets count how many pieces of news for each new type there are

```python
from collections import Counter

# For each piece of news, get it's category
all_categories = map(get_category, data)
# Count them
category_count = Counter(all_categories)
```

Now let's see what keywords appear more often

```python
all_keywords = []
for news in data:
    all_keywords.extend(get_keywords(news))
keyword_count = Counter(all_keywords)
```

We can also see who the most talked about people were

```python
all_people = []
for news in data:
    # Look at each news' entities
    for entity in get_entities(news):
        # If they are a person then save them
        if entity[1] == 'PERSON':
            all_people.append(entity[0])
# Count them
people_count = Counter(all_people)
```

We can also see what the most popular entities were on a pero news type basis

```python
# We will put our results here, keys will be news type, and the values
# will be (counts of people, counts of locations) tuples.
entities_per_topic = {}
# Look at each category - we add them to a set so we don't repeat
for category in set(all_categories):
    all_locations = []
    all_people = []
    # Get the news that are in that category
    for news in filter(lambda n: get_category(n) == category, data):
        # Add the person/location to the correct collection
        for entity in get_entities(news):
            if entity[1] == 'PERSON':
                all_people.append(entity[0])
            elif entity[1] == 'LOCATION':
                all_locations.append(entity[0])
            # We are ignoring other entity types
    # Count them
    entities_per_topic[category] = (Counter(all_people), Counter(all_locations))
```

## Data visualization

We can now take a look at the data we got, and possibly gain some insights

```python
# Import graphing libraries
%matplotlib inline 
import matplotlib.pyplot as plot
import numpy as np

from __future__ import print_function

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
```

```python
# Let's look at the most popular types of news
for category in category_count:
    print('{0}\t{1}'.format(category, category_count[category]))
```

```python
# We can graph it for easier viewing
labels = category_count.keys()
values = [category_count[v] for v in labels]

explode = [0.1] + ([0] * (len(values) -1))

fig_size = plot.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 9
plot.rcParams["figure.figsize"] = fig_size

plot.pie(values, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plot.axis('equal')

plot.show()
```

```python
# Let's look at the most popular people in the news
for person in people_count:
    print('{0}\t{1}'.format(person.encode('UTF-8'), people_count[person]))
```

```python
# We can also graph it for easier viewing
labels = []
values = []

# Lets get the top 10 most talked about people
for l, v in people_count.most_common(10):
    labels.append(l)
    values.append(v)
    
x = np.arange(len(values))

fig_size = plot.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 9
plot.rcParams["figure.figsize"] = fig_size

plot.barh(x, values, color='yellowgreen')
plot.yticks(x + 0.5, labels, rotation=0)
plot.xticks(xrange(max(values) + 1))

plot.show()
```

# Some more comparisons

So far we have only taken a look inside at one of the news outlets. However, it would be much more interesting to compare tham to each other. What we will be doing is loading all news articles and categorizing them. Then we will compare the news outlets.

```python
outlet_data = {}
for outlet in OUTLETS:
    with open(outlet + '.json', 'rb') as f:
        items = json.load(f)['news']

        data = {'text_list': items}
        headers = {'Authorization': 'Token {0}'.format(API_KEY),
                   'Content-Type': 'application/json'}

        response = requests.post(NEWS_CLASSIFIER,
                                 data=json.dumps(data),
                                 headers=headers)
        result = json.loads(response.text)['result']
        
        # For each piece of news, get it's category
        all_categories = map(lambda n: n[-1]['label'], result)
        # Count them
        category_count = Counter(all_categories)
        outlet_data[outlet] = category_count
        
```

Now that we have all the data, let's display it.

```python
# Let's graph each news outlet
for outlet in OUTLETS:
    category_count = outlet_data[outlet]
    
    # The code to display the pie charts is the same tha was used before.
    labels = category_count.keys()
    values = [category_count[v] for v in labels]

    explode = [0.1] + ([0] * (len(values) -1))

    fig_size = plot.rcParams["figure.figsize"]
    fig_size[0] = 20
    fig_size[1] = 9
    plot.rcParams["figure.figsize"] = fig_size

    plot.pie(values, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
    plot.axis('equal')

    print(outlet.replace('_', ' ').upper())
    plot.show()
    
```
